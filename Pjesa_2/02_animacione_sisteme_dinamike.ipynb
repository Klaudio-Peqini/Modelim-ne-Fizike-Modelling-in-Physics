## Importimi i librarive

```

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

## 1. Animacion i një penduli të thjeshtë

```

g = 9.81
L = 1.0

t = np.linspace(0,10,500)
theta = 0.5*np.cos(np.sqrt(g/L)*t)

x = L*np.sin(theta)
y = -L*np.cos(theta)

fig, ax = plt.subplots()
line, = ax.plot([], [], 'o-', lw=2)

ax.set_xlim(-1.2,1.2)
ax.set_ylim(-1.2,0.2)

def update(frame):
    line.set_data([0,x[frame]],[0,y[frame]])
    return line,

ani = FuncAnimation(fig, update, frames=len(t), interval=20)

plt.show()
```

## 2. Animacion i mekanizmit bielë-manivelë

```

r = 1.0
l = 2.5

theta = np.linspace(0,2*np.pi,400)

x_crank = r*np.cos(theta)
y_crank = r*np.sin(theta)

x_piston = r*np.cos(theta) + np.sqrt(l**2 - (r*np.sin(theta))**2)

fig, ax = plt.subplots()

crank_line, = ax.plot([],[],'o-',lw=3)
rod_line, = ax.plot([],[],'o-',lw=3)

ax.set_xlim(-3,3)
ax.set_ylim(-3,3)

def update(frame):

    crank_line.set_data([0,x_crank[frame]],[0,y_crank[frame]])

    rod_line.set_data([x_crank[frame],x_piston[frame]],[y_crank[frame],0])

    return crank_line, rod_line

ani = FuncAnimation(fig, update, frames=len(theta), interval=30)

plt.show()
```

## 3. Animacion i një orbite planetare

```

t = np.linspace(0,20,600)

x = np.cos(t)
y = np.sin(t)

fig, ax = plt.subplots()

planet, = ax.plot([],[],'o')
orbit, = ax.plot(x,y)

ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)

def update(frame):

    planet.set_data(x[frame],y[frame])

    return planet,

ani = FuncAnimation(fig, update, frames=len(t), interval=20)

plt.show()
```

## 4. Spirale në hapësirë fazore

```

t = np.linspace(0,10,500)

x = np.exp(-0.1*t)*np.cos(2*t)
y = np.exp(-0.1*t)*np.sin(2*t)

fig, ax = plt.subplots()

line, = ax.plot([],[])

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)

def update(frame):

    line.set_data(x[:frame],y[:frame])

    return line,

ani = FuncAnimation(fig, update, frames=len(t), interval=20)

plt.show()
```
