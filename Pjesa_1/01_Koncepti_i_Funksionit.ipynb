## 1. Pse na duhet koncepti i funksionit?

Në jetën e përditshme dhe në shkencë, ne hasim vazhdimisht marrëdhënie të tipit:

- **një hyrje** $x$,
- **një rregull** që vepron mbi atë hyrje,
- **një dalje** $y$.

Kjo është thelbi i një funksioni.

Në matematikë shkruajmë, për shembull:

$
f(x) = 2x + 3
$

Kjo do të thotë: merre numrin $x$, shumëzoje me 2, pastaj shto 3.

Pra:

- nëse $x = 1$, atëherë $f(1) = 5$;
- nëse $x = 4$, atëherë $f(4) = 11$;
- nëse $x = -2$, atëherë $f(-2) = -1$.

Në programim, një funksion bën pothuajse të njëjtën gjë:
merr disa hyrje, zbaton një procedurë, dhe kthen një rezultat.

Funksionet janë të nevojshme sepse:

1. **organizojnë mendimin**;
2. **shmangin përsëritjen**;
3. **e bëjnë kodin më të lexueshëm**;
4. **na lejojnë të testojmë pjesë të vogla veçmas**;
5. **e bëjnë një program të ripërdorshëm**.

Në modelim fizik, kjo është thelbësore: formula, ligje, hapa numerikë, përpunime të të dhënave — të gjitha mund të organizohen si funksione.


## 2. Një ide intuitive: funksioni si “makinë”

Një mënyrë shumë e mirë për ta menduar funksionin është si një **makinë**:

- ti fut diçka brenda;
- makina bën një transformim;
- del një rezultat.

Për shembull:

- fut temperaturën në gradë Celsius;
- funksioni e konverton në Kelvin;
- del temperatura në Kelvin.

Matematikisht:

$
T_K(T_C) = T_C + 273.15
$

Në programim, kjo bëhet një funksion.

Le ta shohim fillimisht pa funksion.


````python
T_C = 25
T_K = T_C + 273.15
print("Temperatura në Kelvin është:", T_K)
````

Ky kod funksionon. Por çfarë ndodh nëse duam të konvertojmë 10 temperatura të ndryshme?
Duhet ta përsërisim të njëjtin veprim shumë herë.

Këtu lind nevoja për funksionin.


````python
def celsius_ne_kelvin(T_C):
    return T_C + 273.15

print(celsius_ne_kelvin(25))
print(celsius_ne_kelvin(0))
print(celsius_ne_kelvin(-10))
````

Vini re strukturën:

```python
def celsius_ne_kelvin(T_C):
    return T_C + 273.15
```

Këtu:

- `def` tregon se po **përcaktojmë** një funksion;
- `celsius_ne_kelvin` është **emri** i funksionit;
- `T_C` është **hyrja** ose **parametri**;
- `return` kthen rezultatin.

Pra, funksioni është një bllok kodi i emërtuar që mund ta thërrasim sa herë të duam.

## 4. Çfarë problemi zgjidh funksioni?

Le të shohim një situatë konkrete pa funksion.

Duam të llogarisim energjinë kinetike:

$
E_k = \frac{1}{2} m v^2
$

për disa raste të ndryshme.


````python
m1, v1 = 2.0, 3.0
E1 = 0.5 * m1 * v1**2

m2, v2 = 5.0, 1.5
E2 = 0.5 * m2 * v2**2

m3, v3 = 0.8, 10.0
E3 = 0.5 * m3 * v3**2

print(E1, E2, E3)
````

Ky kod punon, por ka disa probleme:

1. formula përsëritet;
2. nëse gabojmë në një vend, mund të kemi rezultate të pasakta;
3. kodi bëhet i vështirë për t’u mirëmbajtur;
4. nuk është i pastër.

Tani ta organizojmë me funksion.


````python
def energjia_kinetike(m, v):
    return 0.5 * m * v**2

print("Rasti 1:", energjia_kinetike(2.0, 3.0), "J")
print("Rasti 2:", energjia_kinetike(5.0, 1.5), "J")
print("Rasti 3:", energjia_kinetike(0.8, 10.0), "J")
````

Tani kemi:

- një formulë të shkruar vetëm një herë;
- një emër kuptimplotë;
- një objekt që mund të ripërdoret;
- kod më të qartë.

Ky është një nga përfitimet më të mëdha të funksioneve.

## 5. Sintaksa bazë e një funksioni

Forma bazë është:

```python
def emri_funksionit(parametri1, parametri2, ...):
    # trupi i funksionit
    return rezultati
```

### Elementët kryesorë

**1. `def`**  
Shërben për të deklaruar një funksion të ri.

**2. Emri i funksionit**  
Duhet të jetë i qartë dhe domethënës.

**3. Parametrat**  
Janë vlerat hyrëse që funksioni merr.

**4. Trupi i funksionit**  
Janë veprimet që funksioni kryen.

**5. `return`**  
Kthen rezultatin jashtë funksionit.

Le ta shohim me një shembull shumë të thjeshtë.


````python
def dyfishi(x):
    rezultat = 2 * x
    return rezultat

print(dyfishi(4))
````

Këtu:

- `x` është input-i;
- `rezultat = 2 * x` është përpunimi;
- `return rezultat` është output-i.

Pra, funksioni ka hyrje, përpunim, dhe dalje.

## 6. Funksione me një parametër dhe me disa parametra

### 6.1 Një parametër


````python
def kubi(x):
    return x**3

print(kubi(2))
print(kubi(-4))
````

### 6.2 Dy parametra


````python
def fuqia_baze_eksponent(baza, eksponenti):
    return baza**eksponenti

print(fuqia_baze_eksponent(2, 3))
print(fuqia_baze_eksponent(5, 2))
````

### 6.3 Tre parametra
Për shembull, formula e pozicionit në lëvizje me përshpejtim konstant:

$
x(t) = x_0 + v_0 t + \frac{1}{2} a t^2
$


````python
def pozicioni(x0, v0, a, t):
    return x0 + v0*t + 0.5*a*t**2

print(pozicioni(0, 2, 1, 3))
print(pozicioni(5, -1, 0.5, 4))
````

Në fizikë, kjo mënyrë e të menduarit është shumë e fuqishme:
çdo ligj ose marrëdhënie mund të enkapsulohet si funksion.

## 7. Pse nuk mjafton thjesht të shkruajmë kod rresht pas rreshti?

Në programe të vogla, mund të duket sikur funksionet nuk janë të domosdoshme.
Por sapo problemi rritet, kodi linear bëhet:

- i gjatë,
- i çrregullt,
- i vështirë për t’u testuar,
- i vështirë për t’u ripërdorur.

Le të shohim një shembull të thjeshtë: duam të llogarisim sipërfaqen e rrethit për disa rreze.


````python
pi = 3.141592653589793

r1 = 1
A1 = pi * r1**2

r2 = 2.5
A2 = pi * r2**2

r3 = 4
A3 = pi * r3**2

print(A1, A2, A3)
````

Tani me funksion:


````python
def siperfaqja_rrethit(r):
    pi = 3.141592653589793
    return pi * r**2

for r in [1, 2.5, 4]:
    print(f"r = {r}, A = {siperfaqja_rrethit(r)}")
````

Kjo qasje është më elegante dhe më shkallëzueshme.

Në projektet shkencore, funksionet janë një nga mekanizmat kryesorë që e bëjnë kodin **modular**.

## 9. `return` kundrejt `print`

Studentët shpesh i ngatërrojnë këto dy ide.

### `print`
Shërben për të shfaqur diçka në ekran.

### `return`
Shërben për të kthyer një vlerë që mund të përdoret më tej.

Le ta shohim ndryshimin.


````python
def me_print(x):
    print(x**2)

def me_return(x):
    return x**2

a = me_print(3)
b = me_return(3)

print("Vlera e a është:", a)
print("Vlera e b është:", b)
````

Çfarë ndodhi?

- `me_print(3)` vetëm e shfaqi rezultatin;
- `me_return(3)` e ktheu rezultatin;
- prandaj `b` ruajti vlerën `9`, ndërsa `a` mori vlerën `None`.

Ky është një dallim shumë i rëndësishëm.

Në programim shkencor, zakonisht na duhen funksione që **kthejnë** vlera, sepse më pas duam t’i përdorim në llogaritje të tjera.

## 11. Funksione dhe ripërdorim

Një arsye shumë e fortë pse përdorim funksione është **ripërdorimi i kodit**.

Le të themi se na duhet të llogarisim energjinë potenciale gravitacionale:

$
E_p = mgh
$

dhe ta përdorim për shumë objekte të ndryshme.


````python
def energjia_potenciale(m, g, h):
    return m * g * h

rastet = [
    (1.0, 9.81, 2.0),
    (5.0, 9.81, 1.2),
    (0.2, 9.81, 10.0)
]

for m, g, h in rastet:
    print(f"m={m}, g={g}, h={h}  ->  Ep={energjia_potenciale(m, g, h):.3f} J")
````

Pa funksion, do ta shkruanim formulën shumë herë.
Me funksion, e shkruajmë një herë dhe e përdorim sa të duam.

## 13. Një shembull pak më realist: vlera e një polinomi

Le të marrim funksionin:

$
p(x) = x^3 - 2x^2 + 5x - 1
$

Në vend që ta llogarisim manualisht për shumë vlera, ndërtojmë funksion.


````python
def polinomi(x):
    return x**3 - 2*x**2 + 5*x - 1

for x in [-2, -1, 0, 1, 2, 3]:
    print(f"p({x}) = {polinomi(x)}")
````

Ky është një hap shumë i rëndësishëm drejt modelimit numerik:

## 14. Funksione që ndërtohen mbi funksione të tjera

Një tjetër ide shumë e rëndësishme:
mund të përdorim një funksion brenda një funksioni tjetër.

Për shembull:


````python
def katrori(x):
    return x**2

def shuma_e_katroreve(a, b):
    return katrori(a) + katrori(b)

print(shuma_e_katroreve(3, 4))
````

Kjo e bën kodin më modular.

## 17. Një shembull fizik më i plotë: ligji i Ohmit

Ligji i Ohmit:

$
V = IR
$

ku:

- $V$ është tensioni,
- $I$ rryma,
- $R$ rezistenca.

Mund të shkruajmë një funksion që llogarit tensionin.


````python
def tensioni(I, R):
    return I * R

print("V =", tensioni(2, 5), "V")
print("V =", tensioni(0.1, 220), "V")
````

Dhe mund të shkruajmë edhe funksione të tjera nga e njëjta marrëdhënie.


````python
def rryma(V, R):
    return V / R

def rezistenca(V, I):
    return V / I

print("I =", rryma(12, 6), "A")
print("R =", rezistenca(12, 2), "ohm")
````

Këtu po shihni diçka shumë të rëndësishme:
**një teori fizike mund të organizohet si një grup funksionesh**.

- dhe më pas programe më të strukturuara.

## 23. Detyrë e vogël shtëpie

Ndërtoni tre funksione tuajat:

1. një funksion nga matematika;
2. një funksion nga fizika;
3. një funksion nga jeta e përditshme.

Për secilin:
- shkruani formulën;
- shkruani funksionin në Python;
- testojeni për disa vlera;
- shpjegoni shkurt se pse ky funksion është i dobishëm.

Ky ushtrim synon t’ju ndihmojë të kaloni nga **të kuptuarit teorik** te **zbatimi praktik**.
