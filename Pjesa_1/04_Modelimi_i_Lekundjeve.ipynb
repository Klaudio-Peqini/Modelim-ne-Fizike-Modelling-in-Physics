## 2. Sistemi model: masa–sustë

Modeli më i thjeshtë është një masë $m$ e lidhur me një sustë me konstante elastike $k$.

Nëse masa zhvendoset nga pozicioni i ekuilibrit me një shumësi $x$, susta ushtron forcën:

$
F = -kx
$

Shenja minus tregon që forca është **rikthyese**: ajo drejtohet gjithmonë kundër zhvendosjes.

Sipas ligjit të dytë të Njutonit:

$
m \frac{d^2x}{dt^2} = -kx
$

ose:

$
\frac{d^2x}{dt^2} + \frac{k}{m}x = 0
$

Ky është ekuacioni diferencial i **lëkundjes harmonike të thjeshtë**.


### Interpretimi i parametrave

Në këtë model:

- $x(t)$ është zhvendosja në kohë;
- $m$ është masa;
- $k$ është ngurtësia e sustës.

Sa më i madh të jetë $k$, aq më e fortë është prirja rikthyese.  
Sa më e madhe të jetë $m$, aq më “i rëndë” është sistemi dhe aq më ngadalë reagon ai.

## 3. Nga ekuacioni diferencial te zgjidhja

Ekuacioni:

$
\frac{d^2x}{dt^2} + \omega_0^2 x = 0
$

me

$
\omega_0 = \sqrt{\frac{k}{m}}
$

ka zgjidhjen e përgjithshme:

$
x(t) = A \cos(\omega_0 t + \phi)
$

ku:

- $A$ është amplituda;
- $\omega_0$ është frekuenca këndore natyrore;
- $\phi$ është faza fillestare.

Pra, sjellja e sistemit është sinusoidale dhe periodike.


```
import math

def lekundja_harmonike(A, omega, phi, t):
    return A * math.cos(omega * t + phi)

A = 2.0
omega = 3.0
phi = 0.0

for t in [0, 0.5, 1.0, 1.5, 2.0]:
    x = lekundja_harmonike(A, omega, phi, t)
    print(f"t = {t:>3.1f} s   ->   x = {x:>7.4f}")
```

Këtu po përdorim drejtpërdrejt zgjidhjen analitike.

Kjo është shumë e rëndësishme didaktikisht: studenti sheh menjëherë se ekuacioni diferencial nuk është vetëm simbolik, por përshkruan një lëvizje konkrete, të matshme dhe të llogaritshme.

## 4. Perioda dhe frekuenca

Për lëkundjen harmonike të thjeshtë kemi:

$
T = \frac{2\pi}{\omega_0}
$

dhe

$
f = \frac{1}{T}
$

ku:

- $T$ është perioda;
- $f$ është frekuenca.

Le të shkruajmë funksione të vogla për këto madhësi.


```
def omega_natyrore(k, m):
    return math.sqrt(k / m)

def perioda(omega):
    return 2 * math.pi / omega

def frekuenca(omega):
    return omega / (2 * math.pi)

k = 9.0
m = 1.0
omega0 = omega_natyrore(k, m)

print("omega0 =", omega0)
print("T =", perioda(omega0))
print("f =", frekuenca(omega0))
```

Kjo na tregon një ide fizike shumë të rëndësishme:

- sistemi “e ka” një ritëm natyror të vetin;
- ky ritëm varet vetëm nga parametrat e modelit;
- ndryshimi i $k$ ose $m$ ndryshon karakterin kohor të lëkundjes.

## 5. Kushtet fillestare dhe faza

Zgjidhja e lëkundjes varet nga kushtet fillestare.

Për shembull:

- nëse në $t=0$ sistemi nis nga zhvendosja maksimale me shpejtësi zero, shpesh kemi formën me kosinus;
- nëse nis nga ekuilibri me shpejtësi fillestare jo-zero, forma mund të shfaqet më natyrshëm me sinus.

Pra, **amplituda** dhe **faza** kodojnë informacionin mbi gjendjen fillestare.

Le të krahasojmë dy raste me të njëjtën amplitudë dhe frekuencë, por me faza të ndryshme.


```
A = 1.0
omega = 2.0

for phi in [0.0, math.pi/4, math.pi/2]:
    print(f"--- faza phi = {phi:.3f} rad ---")
    for t in [0, 0.5, 1.0]:
        print(f"t={t}, x={lekundja_harmonike(A, omega, phi, t):.4f}")
    print()
```

Kjo tregon se faza nuk ndryshon “natyrën” e lëkundjes, por vetëm pozicionin e saj kohor fillestar.

## 6. Shpejtësia dhe energjia në lëkundjen harmonike

Nëse:

$
x(t) = A \cos(\omega t + \phi)
$

atëherë shpejtësia është:

$
v(t) = -A\omega \sin(\omega t + \phi)
$

Ndërsa energjia totale e sistemit ideal është konstante:

$
E = \frac{1}{2}mv^2 + \frac{1}{2}kx^2
$

Kjo është një nga veçoritë më të rëndësishme të sistemit ideal: nuk ka humbje energjie.


```
def shpejtesia_harmonike(A, omega, phi, t):
    return -A * omega * math.sin(omega * t + phi)

def energjia_totale(m, k, x, v):
    return 0.5 * m * v**2 + 0.5 * k * x**2

A = 1.5
k = 4.0
m = 1.0
omega = math.sqrt(k/m)
phi = 0.0

for t in [0, 0.3, 0.6, 0.9]:
    x = lekundja_harmonike(A, omega, phi, t)
    v = shpejtesia_harmonike(A, omega, phi, t)
    E = energjia_totale(m, k, x, v)
    print(f"t={t:.1f}, x={x:.4f}, v={v:.4f}, E={E:.4f}")
```

Edhe pse $x$ dhe $v$ ndryshojnë me kohën, energjia totale mbetet e pandryshuar në modelin ideal.

Ky është një shembull shumë i bukur i lidhjes midis:
- ligjeve të lëvizjes,
- zgjidhjes analitike,
- dhe ruajtjes së energjisë.

## 7. Tabela e evolucionit kohor

Një hap i natyrshëm në modelim është të ndërtojmë jo vetëm një vlerë të vetme, por një evolucion të plotë në kohë.


```
def tabela_harmonike(A, omega, phi, t_max, dt):
    times = []
    positions = []

    t = 0.0
    while t <= t_max:
        times.append(t)
        positions.append(lekundja_harmonike(A, omega, phi, t))
        t += dt

    return times, positions

times, positions = tabela_harmonike(1.0, 2.0, 0.0, 4.0, 0.5)

for t, x in zip(times, positions):
    print(f"t={t:>3.1f}   x={x:>7.4f}")
```

Kjo është qasja tipike llogaritëse:
- caktojmë parametrat;
- caktojmë intervalin kohor;
- ndërtojmë tabelën e rezultateve;
- më pas mund ta vizatojmë ose analizojmë më tej.

## 8. Lëkundja e shuarur

Në sistemet reale, zakonisht ka humbje energjie për shkak të fërkimit, rezistencës së ajrit, disipimit elektrik, etj.

Atëherë modeli bëhet:

$
m\frac{d^2x}{dt^2} + b\frac{dx}{dt} + kx = 0
$

ku $b$ është koeficienti i shuarjes.

Ky term shtesë është shumë i rëndësishëm fizikisht: ai përfaqëson mekanizmin e humbjes së energjisë.


### Forma e zgjidhjes së shuarur

Në rastin e nënshuarjes (underdamped), zgjidhja ka formën:

$
x(t) = A e^{-\gamma t} \cos(\omega_d t + \phi)
$

ku:

- $\gamma$ kontrollon shuarjen;
- $\omega_d$ është frekuenca e shuarur.

Pra, sistemi ende lëkundet, por amplituda zvogëlohet me kohën.


```
def lekundja_e_shuarur(A, gamma, omega_d, phi, t):
    return A * math.exp(-gamma * t) * math.cos(omega_d * t + phi)

A = 2.0
gamma = 0.25
omega_d = 2.5
phi = 0.0

for t in [0, 1, 2, 3, 4, 5]:
    x = lekundja_e_shuarur(A, gamma, omega_d, phi, t)
    print(f"t={t}, x={x:.4f}")
```

Në këtë model amplituda bie si eksponenciale, ndërsa pjesa kosinusoidale përshkruan lëkundjen.

Kjo është një lidhje shumë e bukur midis dy temave të mëparshme:
- prishjes eksponenciale;
- lëkundjes harmonike.

## 9. Krahasimi: ideal kundrejt i shuarur

Le të krahasojmë të njëjtin sistem pa shuarje dhe me shuarje.


```
A = 1.0
omega = 2.0
gamma = 0.2
omega_d = 1.9
phi = 0.0

print("t      ideal        shuarur")
for t in [0, 0.5, 1.0, 1.5, 2.0, 2.5]:
    x_ideal = lekundja_harmonike(A, omega, phi, t)
    x_shuarur = lekundja_e_shuarur(A, gamma, omega_d, phi, t)
    print(f"{t:<4.1f}   {x_ideal:>8.4f}   {x_shuarur:>8.4f}")
```

Krahasimi tregon qartë:

- sistemi ideal ruan amplitudën;
- sistemi i shuarur e humbet gradualisht atë.

Ky është një hap shumë i rëndësishëm konceptual: një model më realist zakonisht është edhe më i pasur fizikisht.

## 10. Regjimet e shuarjes

Sjellja e sistemit të shuarur varet nga balanca midis inercisë, shuarjes dhe forcës rikthyese.

Në përgjithësi dallohen tre regjime:

1. **Nënshuarja (underdamped)**  
   sistemi lëkundet dhe amplituda bie.

2. **Shuarja kritike (critically damped)**  
   sistemi kthehet sa më shpejt në ekuilibër pa lëkundje.

3. **Mbishuarja (overdamped)**  
   sistemi kthehet në ekuilibër pa lëkundje, por më ngadalë se rasti kritik.

Ky klasifikim është shumë i rëndësishëm jo vetëm në mekanikë, por edhe në elektronikë, kontroll automatik dhe shumë fusha të tjera.

## 11. Lëkundja e detyruar

Tani marrim një sistem që përveç forcës rikthyese dhe shuarjes, ndikohet edhe nga një forcë e jashtme periodike:

$
m\frac{d^2x}{dt^2} + b\frac{dx}{dt} + kx = F_0 \cos(\Omega t)
$

Këtu:
- $F_0$ është amplituda e forcës së jashtme;
- $\Omega$ është frekuenca këndore e ngacmimit.

Ky model quhet **lëkundje e detyruar**.


### Pse është i rëndësishëm ky model?

Sepse në botën reale shumë sisteme nuk janë të izoluara.  
Ato ndikohen nga mjedisi, nga burime të jashtme, nga ngacmime periodike.

Shembuj:
- një urë e ngacmuar nga era;
- një qark elektrik i ushqyer me sinjal periodik;
- një strukturë mekanike e ngacmuar nga vibrime;
- sisteme klimatike me cikle periodike të jashtme.

## 12. Rezonanca

Një nga fenomenet më të rëndësishme në fizikën e lëkundjeve është **rezonanca**.

Kur frekuenca e forcës së jashtme afrohet me frekuencën natyrore të sistemit, amplituda e përgjigjes mund të rritet shumë.

Pra:

> sistemi “përgjigjet” më fort kur ngacmohet pranë ritmit të vet natyror.

Kjo ide ka rëndësi të jashtëzakonshme në:
- inxhinieri;
- akustikë;
- elektronikë;
- fizikë atomike;
- dinamika strukturore.


```
def amplituda_rezonances(F0, m, omega0, Omega, beta):
    denominator = math.sqrt((omega0**2 - Omega**2)**2 + (2*beta*Omega)**2)
    return (F0 / m) / denominator

F0 = 1.0
m = 1.0
omega0 = 2.0
beta = 0.15

for Omega in [0.5, 1.0, 1.5, 1.8, 2.0, 2.2, 2.5, 3.0]:
    A = amplituda_rezonances(F0, m, omega0, Omega, beta)
    print(f"Omega={Omega:.1f}   ->   A={A:.4f}")
```

Edhe pa bërë grafik, tabela tregon se amplituda rritet ndjeshëm kur $\Omega$ afrohet pranë $\omega_0$.

Kjo është thelbi i rezonancës.

## 13. Hapësira e fazës

Një mënyrë shumë e rëndësishme për të parë sistemet dinamike është jo vetëm përmes $x(t)$, por edhe përmes çiftit:

$
(x, v)
$

Kjo quhet **hapësira e fazës**.

Për lëkundjen harmonike ideale:
- trajektorja në hapësirën e fazës është eliptike;
- sistemi qarkullon vazhdimisht pa humbje energjie.

Për lëkundjen e shuarur:
- trajektorja spiralet drejt origjinës;
- sistemi humbet energji dhe përfundon në ekuilibër.


```
A = 1.0
omega = 2.0
phi = 0.0

print("t      x(t)       v(t)")
for t in [0, 0.4, 0.8, 1.2, 1.6]:
    x = lekundja_harmonike(A, omega, phi, t)
    v = shpejtesia_harmonike(A, omega, phi, t)
    print(f"{t:<4.1f}   {x:>7.4f}   {v:>7.4f}")
```

Kjo ide është shumë e rëndësishme për kurset e mëvonshme mbi dinamiken, sistemet jolineare dhe kaosin.

## 14. Nga zgjidhja analitike te zgjidhja numerike

Deri tani kemi përdorur formula analitike për disa raste të njohura.

Por shumë sisteme:
- nuk kanë zgjidhje të thjeshtë të mbyllur;
- ose kanë zgjidhje shumë të ndërlikuar;
- ose përfshijnë terma jolinearë.

Atëherë kalojmë te metoda numerike.

Për shembull, ekuacioni i lëkundjes harmonike mund të shkruhet si sistem rendi të parë:

$
\frac{dx}{dt} = v
$

$
\frac{dv}{dt} = -\omega_0^2 x
$

Kjo është forma ideale për integrim numerik.


```
def derivatet_harmonike(x, v, omega0):
    dxdt = v
    dvdt = -(omega0**2) * x
    return dxdt, dvdt

x = 1.0
v = 0.0
omega0 = 2.0

dxdt, dvdt = derivatet_harmonike(x, v, omega0)
print("dx/dt =", dxdt)
print("dv/dt =", dvdt)
```

Kjo lidh drejtpërdrejt këtë notebook me notebook-un për **paradigmat dhe metodat ODE**.

## 15. Një hap i thjeshtë Euler për lëkundjen harmonike

Për qëllime didaktike, le të bëjmë disa hapa shumë të thjeshtë me metodën Euler.

**Kujdes:** kjo metodë nuk është gjithmonë ideale për lëkundje afatgjata, por është shumë e dobishme për të kuptuar idenë e integrimit numerik.


```
def euler_harmonik(x0, v0, omega0, dt, n_steps):
    times = [0.0]
    xs = [x0]
    vs = [v0]

    x = x0
    v = v0
    t = 0.0

    for _ in range(n_steps):
        dxdt = v
        dvdt = -(omega0**2) * x

        x = x + dt * dxdt
        v = v + dt * dvdt
        t = t + dt

        times.append(t)
        xs.append(x)
        vs.append(v)

    return times, xs, vs

times, xs, vs = euler_harmonik(1.0, 0.0, 2.0, 0.05, 10)

for t, x, v in zip(times, xs, vs):
    print(f"t={t:.2f}, x={x:.4f}, v={v:.4f}")
```

Kjo është një nga pikat më të rëndësishme të të gjithë Bundle 1:

- fillojmë me ide fizike;
- shkruajmë ekuacionin diferencial;
- e riformulojmë si sistem;
- pastaj e integrojmë numerikisht.

Pra, fizikë → model → ekuacion → algoritëm.

Një pjesë e rëndësishme e modelimit është të kuptojmë jo vetëm formulat, por edhe **rolin fizik të çdo parametri**.


```
m = 1.0
for k in [1.0, 4.0, 9.0, 16.0]:
    omega0 = omega_natyrore(k, m)
    T = perioda(omega0)
    print(f"k={k:>4.1f}   omega0={omega0:>6.3f}   T={T:>6.3f}")
```
