# Pjesa 3 – Oshilatore Jo‑lineare dhe Modifikime të Oshilatorit Harmonike

Ky notebook është pjesë e kursit **Modelling in Physics** dhe prezanton disa modifikime të rëndësishme të oshilatorit harmonik:

- Oshilatori anharmonik
- Oshilatori Duffing
- Oshilatori van der Pol
- Oshilatorë të lidhur shumëdimensionale

Qëllimi është të tregojmë se si modifikimet jo‑lineare prodhojnë dinamika shumë më të pasura.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

## Oshilatori harmonik (rikujtim)

Ekuacioni i lëvizjes:

$$ m\ddot{x}+kx=0 $$

ose

$$ \ddot{x}+\omega^2 x = 0 $$

```python
def harmonic_oscillator(t, y, omega):
    x, v = y
    dxdt = v
    dvdt = -omega**2 * x
    return [dxdt, dvdt]

t_span = (0, 30)
t_eval = np.linspace(*t_span, 1000)

sol = solve_ivp(harmonic_oscillator, t_span, [1,0], t_eval=t_eval, args=(1.0,))

plt.plot(sol.t, sol.y[0])
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Oshilatori harmonik')
plt.show()
```

## Oshilatori Anharmonik

Potenciali:

$$V(x)=\frac{1}{2}kx^2+\alpha x^4$$

Ekuacioni i lëvizjes:

$$ m\ddot{x}+kx+4\alpha x^3=0 $$

```python
def anharmonic_oscillator(t, y, omega, alpha):
    x, v = y
    dxdt = v
    dvdt = -omega**2 * x - alpha * x**3
    return [dxdt, dvdt]

sol = solve_ivp(anharmonic_oscillator,(0,40),[1,0],t_eval=np.linspace(0,40,2000),args=(1.0,0.5))

plt.plot(sol.t, sol.y[0])
plt.title('Oshilatori anharmonik')
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

## Oshilatori Duffing

$$ \ddot{x}+\delta\dot{x}+\alpha x+\beta x^3=\gamma\cos(\omega t) $$

```python
def duffing(t, y, delta, alpha, beta, gamma, omega):
    x, v = y
    dxdt = v
    dvdt = -delta*v - alpha*x - beta*x**3 + gamma*np.cos(omega*t)
    return [dxdt, dvdt]

params = (0.2,-1,1,0.3,1.2)

sol = solve_ivp(duffing,(0,100),[0.5,0],t_eval=np.linspace(0,100,4000),args=params)

plt.plot(sol.t, sol.y[0])
plt.title('Duffing oscillator')
plt.show()
```

## Oshilatori van der Pol

$$ \ddot{x}-\mu(1-x^2)\dot{x}+x=0 $$

```python
def vdp(t, y, mu):
    x, v = y
    dxdt = v
    dvdt = mu*(1-x**2)*v - x
    return [dxdt, dvdt]

sol = solve_ivp(vdp,(0,60),[2,0],t_eval=np.linspace(0,60,4000),args=(3,))

plt.plot(sol.t, sol.y[0])
plt.title('van der Pol oscillator')
plt.show()
```

## Oshilatorë të lidhur (2D)

$$
\ddot{x}+\omega_x^2 x+\epsilon y=0
$$

$$
\ddot{y}+\omega_y^2 y+\epsilon x=0
$$

```python
def coupled_oscillators(t, y, wx, wy, eps):
    x, vx, y_, vy = y

    dxdt = vx
    dvxdt = -wx**2 * x - eps*y_

    dydt = vy
    dvydt = -wy**2 * y_ - eps*x

    return [dxdt, dvxdt, dydt, dvydt]

sol = solve_ivp(coupled_oscillators,(0,50),[1,0,0.5,0],t_eval=np.linspace(0,50,2000),args=(1,1.2,0.3))

plt.plot(sol.t, sol.y[0], label='x')
plt.plot(sol.t, sol.y[2], label='y')

plt.legend()
plt.title('Coupled oscillators')
plt.show()
```
